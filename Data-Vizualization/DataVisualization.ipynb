# Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset into a pandas DataFrame
df = pd.read_csv('synthetic_dataset_with_risk_lakh2.csv')

# Filter the dataset for a particular state and substation
state = 'Kerala'  # Replace 'Kerala' with the actual state name
substation = 'TRISSUR HVDC'  # Replace 'TRISSUR HVDC' with the name of the substation
filtered_df = df[(df['State'] == state)]

# Extract columns for temperature, wind, humidity, rainfall, and pressure
temperature = filtered_df['Temp_K']
wind = filtered_df['wind_speed_m_sec']
humidity = filtered_df['humidity_percent']
rainfall = filtered_df['rain_volume_mm']
pressure = filtered_df['pressure_hPa']

# Create histograms for temperature, wind, humidity, rainfall, and pressure
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.hist(temperature, bins=20, color='#1f77b4', alpha=0.7, width=0.8)  # Dark blue
plt.xlabel('Temperature (K)')
plt.ylabel('Frequency')
plt.title('Temperature Histogram')

plt.subplot(2, 3, 2)
plt.hist(wind, bins=20, color='#2ca02c', alpha=0.7, width=0.8)  # Dark green
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Frequency')
plt.title('Wind Speed Histogram')

plt.subplot(2, 3, 3)
plt.hist(humidity, bins=20, color='#ff7f0e', alpha=0.7, width=0.8)  # Dark orange
plt.xlabel('Humidity (%)')
plt.ylabel('Frequency')
plt.title('Humidity Histogram')

plt.subplot(2, 3, 4)
plt.hist(rainfall, bins=20, color='#d62728', alpha=0.7, width=0.8)  # Dark red
plt.xlabel('Rainfall (mm)')
plt.ylabel('Frequency')
plt.title('Rainfall Histogram')

plt.subplot(2, 3, 5)
plt.hist(pressure, bins=20, color='#9467bd', alpha=0.7, width=0.8)  # Dark purple
plt.xlabel('Pressure (hPa)')
plt.ylabel('Frequency')
plt.title('Pressure Histogram')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset into a pandas DataFrame
df = pd.read_csv('synthetic_dataset_with_risk_lakh2.csv')

# Filter the dataset for a particular state and substation
state = 'Kerala'  # Replace 'Kerala' with the actual state name
substation = 'TRISSUR HVDC'  # Replace 'TRISSUR HVDC' with the name of the substation
filtered_df = df[(df['State'] == state)]

# Extract the 'Risk' column
risk = filtered_df['Risk']

# Count the occurrences of each risk level
risk_counts = risk.value_counts()

# Customize the labels
risk_labels = ['Low', 'High']

# Plot a pie chart for the distribution of risk levels
plt.figure(figsize=(8, 8))
plt.pie(risk_counts, labels=risk_labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Risk Levels')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset into a pandas DataFrame
df = pd.read_csv('synthetic_dataset_with_risk_lakh2.csv')

# Filter the dataset for a particular state and substation
state = 'Kerala'  # Replace 'Kerala' with the actual state name
substation = 'TRISSUR HVDC'  # Replace 'TRISSUR HVDC' with the name of the substation
filtered_df = df[(df['State'] == state)]

# Select only the first 10,000 rows
filtered_df = filtered_df.head(100)

# Extract columns for temperature, wind, humidity, rainfall, and pressure
temperature = filtered_df['Temp_K']
wind = filtered_df['wind_speed_m_sec']
humidity = filtered_df['humidity_percent']
rainfall = filtered_df['rain_volume_mm']
pressure = filtered_df['pressure_hPa']

plt.figure(figsize=(12, 8))

# Scatter plot for Temperature vs. Frequency
plt.subplot(2, 3, 1)
plt.scatter(temperature, range(len(temperature)), color='blue', alpha=0.7, marker='o', s=50, label='Temperature')
plt.xlabel('Temperature (K)')
plt.ylabel('Data Point')
plt.title('Temperature vs. Data Point')
plt.legend()

# Scatter plot for Wind Speed vs. Frequency
plt.subplot(2, 3, 2)
plt.scatter(wind, range(len(wind)), color='green', alpha=0.7, marker='s', s=50, label='Wind Speed')
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Data Point')
plt.title('Wind Speed vs. Data Point')
plt.legend()

# Scatter plot for Humidity vs. Frequency
plt.subplot(2, 3, 3)
plt.scatter(humidity, range(len(humidity)), color='orange', alpha=0.7, marker='^', s=50, label='Humidity')
plt.xlabel('Humidity (%)')
plt.ylabel('Data Point')
plt.title('Humidity vs. Data Point')
plt.legend()

# Scatter plot for Rainfall vs. Frequency
plt.subplot(2, 3, 4)
plt.scatter(rainfall, range(len(rainfall)), color='red', alpha=0.7, marker='d', s=50, label='Rainfall')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Data Point')
plt.title('Rainfall vs. Data Point')
plt.legend()

# Scatter plot for Pressure vs. Frequency
plt.subplot(2, 3, 5)
plt.scatter(pressure, range(len(pressure)), color='purple', alpha=0.7, marker='*', s=50, label='Pressure')
plt.xlabel('Pressure (hPa)')
plt.ylabel('Data Point')
plt.title('Pressure vs. Data Point')
plt.legend()

# Gauge chart
plt.subplot(2, 3, 6)
labels = ['Temperature', 'Wind Speed', 'Humidity', 'Rainfall', 'Pressure']
sizes = [len(temperature)+7, len(wind)+2, len(humidity)-8, len(rainfall)-4, len(pressure)+3]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Data Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.tight_layout()
plt.show()
```
